# IMAGE WITH MATRICES

```python
import numpy as np
import matplotlib.pyplot as plt
```

### saving array as file

```python
array_1 = np.array([[1,2,3],
                    [4,5,6]])
array_2 = np.array([[1,2],
                    [3,4]])
array_3 = np.random.rand(3,3)

np.save('array_1.npy',array_1)
np.save('array_2.npy',array_2)
np.save('array_3.npy',array_3)

loaded_array1 = np.load('array_1.npy')
loaded_array2 = np.load('array_2.npy')
loaded_array3 = np.load('array_3.npy')

print(loaded_array1)
print(loaded_array2)
print(loaded_array3)
```

## Image Displaying

```python
try: 
    logo = np.load('numpy-logo.npy')
    
    #display
    plt.figure(figsize = (10,5))
    plt.subplot(1,2,1)
    plt.imshow(logo)
    plt.title("numpy logo")
    plt.grid(False)

    dark_logo = 1 - logo # DARK IMAGE LOGIC (INVERT COLOR)

    plt.figure(figsize = (10,5))
    plt.subplot(1,2,1)
    plt.imshow(dark_logo)
    plt.title("numpy Dark logo")
    plt.grid(False)


except FileNotFoundError:
    print("numpy logo file not found")
```
